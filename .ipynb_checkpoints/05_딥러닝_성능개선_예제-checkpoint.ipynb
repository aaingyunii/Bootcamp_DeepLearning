```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, optimizers

import numpy as np
import matplotlib.pyplot as plt
import random

random.seed(0)
np.random.seed(0)
tf.random.set_seed(0)
```

```python
def plot_result(hist, loss_min=None, loss_max=None):
    
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.plot(hist.epoch, hist.history['loss'], label="train")
    plt.plot(hist.epoch, hist.history['val_loss'], label='validation')
    plt.title('Loss')
    if loss_min!=None and loss_max!=None:
        plt.ylim(loss_min, loss_max)
        
    plt.xlabel('에폭수')
    plt.ylabel('loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(hist.epoch, hist.history['accuracy'], label='train')
    plt.plot(hist.epoch, hist.history['val_accuracy'], label='validation')
    plt.title('Accuracy')
    plt.xlabel('에폭수')
    plt.ylabel('accuracy')
    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
# MNIST 데이터셋 로딩
(train_image, train_label), (test_image, test_label)  = keras.datasets.mnist.load_data()

# 전처리 - X: scaling, y: one hot encoding 처리
y_train = keras.utils.to_categorical(train_label)
y_test = keras.utils.to_categorical(test_label)

X_train = train_image.astype('float32')/255
X_test = test_image.astype('float32')/255

print(y_train.shape, y_test.shape)
print(X_train.dtype, X_test.dtype)
```

```python
# 하이퍼파리미터
LEARNING_RATE = 0.001
N_EPOCH = 20
N_BATCH = 1000
```

```python
# Dataset 생성
train_dataset = tf.data.Dataset.from_tensor_slices((X_train, y_train))\
                               .shuffle(60000).batch(N_BATCH, drop_remainder=True)
test_dataset = tf.data.Dataset.from_tensor_slices((X_test, y_test)).batch(N_BATCH)

print(len(train_dataset), len(test_dataset))
```

# 모델의 크기 변경 -> 성능 확인

```python
def get_small_model():
    model = keras.Sequential()
    model.add(layers.Flatten(input_shape=(28,28))) 
    model.add(layers.Dense(10, activation='softmax', name='output_layer'))
    
    model.compile(optimizer=optimizers.Adam(learning_rate=LEARNING_RATE), 
                  loss='categorical_crossentropy', metrics=['accuracy'])
    return model    
```

```python
small_model = get_small_model()
small_model.summary()
```

```python
hist = small_model.fit(train_dataset, epochs=N_EPOCH, validation_data=test_dataset)
```

```python
# 시각화
# plot_result(hist)
plot_result(hist, 0.25, 0.5)
```

```python
def get_big_model():
    model = keras.Sequential()
    model.add(layers.Flatten(input_shape=(28,28)))
    
    model.add(layers.Dense(512, activation='relu'))
    model.add(layers.Dense(512, activation='relu'))
    model.add(layers.Dense(256, activation='relu'))
    model.add(layers.Dense(256, activation='relu'))
    model.add(layers.Dense(128, activation='relu'))
    model.add(layers.Dense(128, activation='relu'))
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(64, activation='relu'))
    
    model.add(layers.Dense(10, activation='softmax', name='output_layer'))
    
    model.compile(optimizer=optimizers.Adam(learning_rate=LEARNING_RATE), 
                  loss='categorical_crossentropy', metrics=['accuracy'])
    return model    
```

```python
big_model = get_big_model()
big_model.summary()
```

```python
hist = big_model.fit(train_dataset, epochs=N_EPOCH, validation_data=test_dataset)
```

```python
# plot_result(hist)
plot_result(hist, 0.05, 0.15)
```

```python
print(f"작은 모델 결과값 [loss, accuracy]: {small_model.evaluate(test_dataset)}")
print(f"큰 모델 결과값 [loss, accuracy] : {big_model.evaluate(test_dataset)}")
```

# Dropout Layer

```python
# Dropout Layer는 Dropout을 적용하려는 Layer 앞에 정의한다.
def get_dropout_model():
    model = keras.Sequential()
    model.add(layers.Flatten(input_shape=(28,28)))
    
    model.add(layers.Dropout(rate=0.3)) # rate 지정, 어느 정도의 비율을 학습에서 제외시킬 지.
    model.add(layers.Dense(512, activation='relu')) # Dropout layer 과 Dense layer는 한 쌍으로 사용됨.
    
    model.add(layers.Dropout(0.3))
    model.add(layers.Dense(512, activation='relu')) 
    
    model.add(layers.Dropout(0.3))
    model.add(layers.Dense(256, activation='relu'))    
    
    model.add(layers.Dropout(0.3))
    model.add(layers.Dense(256, activation='relu'))   
    
    model.add(layers.Dropout(0.3))
    model.add(layers.Dense(128, activation='relu'))    
    
    model.add(layers.Dropout(0.3))
    model.add(layers.Dense(128, activation='relu'))
    
    model.add(layers.Dropout(0.3))
    model.add(layers.Dense(64, activation='relu'))    
    
    model.add(layers.Dropout(0.3))
    model.add(layers.Dense(64, activation='relu'))
    
    model.add(layers.Dropout(0.3))
    model.add(layers.Dense(10, activation='softmax', name='output_layer')) # output layer
    
    model.compile(optimizer=optimizers.Adam(learning_rate=LEARNING_RATE), 
                  loss='categorical_crossentropy', # 원 핫 인코딩은 이미 위에서 실행함.
                  metrics=['accuracy'])
    return model    
```

```python
dropout_model = get_dropout_model()
dropout_model.summary()

## dropout layer는 학습하거나 무언가를 하는 것이 아닌 뒤의 layer의 학습 대상을 제외시키는 역할
### 파라미터가 0이다.
```

```python
# Dropout Layer를 사용하면 사용하지 않았을 때 보다 epoch을 더 늘려야 한다.
hist = dropout_model.fit(train_dataset, epochs=N_EPOCH, validation_data=test_dataset)
```

```python
# 위의 학습 모델을 3번 돌린 결과. (epoch = 20 * 3 = 60)
plot_result(hist)

## Validation의 결과가 계속 좋아지면서 Train 또한 너무 급하지 않게, Validation과 유사하게 
### 좋아지고 있음을 알 수 있다. Overfitting이 많이 나아졌음.
```

# Batch Normalization 예제

```python
def get_model_bn(lr=0.01):
    
    model = keras.Sequential()
    model.add(layers.Flatten(input_shape=(28,28)))
    
    
    # (Dropout Layer ->) Dense Layer -> Batch Normalization -> ReLU(Activation)
    model.add(layers.Dense(256)) # wX+b
    model.add(layers.BatchNormalization()) # 평균 : 0+beta,  표준편차 : 1+gamma
    model.add(layers.ReLU()) # activation function
    
    model.add(layers.Dense(256))
    model.add(layers.BatchNormalization())
    model.add(layers.ReLU())
    
    model.add(layers.Dense(128))
    model.add(layers.BatchNormalization())
    model.add(layers.ReLU())
    
    model.add(layers.Dense(128))
    model.add(layers.BatchNormalization())
    model.add(layers.ReLU())
    
    model.add(layers.Dense(64))
    model.add(layers.BatchNormalization())
    model.add(layers.ReLU())
    
    model.add(layers.Dense(64))
    model.add(layers.BatchNormalization())
    model.add(layers.ReLU())
    
    # output layer
    model.add(layers.Dense(10))
    model.add(layers.Softmax())
#     model.add(layers.Dense(10, activation='softmax'))
    
    model.compile(optimizer=optimizers.Adam(learning_rate=lr), 
                  loss='categorical_crossentropy', metrics=['accuracy'])
    return model    
```

```python
model_bn = get_model_bn(LEARNING_RATE)
model_bn.summary()
```

```python
hist = model_bn.fit(train_dataset, epochs=N_EPOCH, validation_data=test_dataset)
```

```python
plot_result(hist)
```

# Learning Rate 변화를 통한 성능개선

## Learning Rate scheduler 를 이용 (Learning rate Decay)
- 특정 step 마다 learning rate를 변경

```python
len(train_dataset)
```

```python
from tensorflow.keras import optimizers # Optimizer(Adam, RMSProp,...) 들과
                                        # + Learning rate 관련 알고리즘들이 구현된 모듈

lr_scheduler = optimizers.schedules.ExponentialDecay(
    initial_learning_rate=LEARNING_RATE, # 시작 Learning rate
    decay_steps=len(train_dataset)*5,     # Learning rate 를 몇 step 당 변경할 것인지 step 수
    decay_rate=0.5,             # 현재 Learning rate * decay_rate(LR 변경 시 현재 LR에 곱해줄 값)
    staircase=True   # 계단형식으로 변경할 지 여부 (True : 계단형식, False : 매 step 마다 조금씩 변경)
)
```

```python
def get_model_lr(initial_lr=0.01):
    model = keras.Sequential()
    model.add(layers.Flatten(input_shape=(28,28)))
    model.add(layers.Dense(256, activation='relu'))
    model.add(layers.Dense(128, activation='relu'))
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(10, activation='softmax', name='output_layer'))
    
    lr_sch = optimizers.schedules.ExponentialDecay(initial_learning_rate=initial_lr, 
                                                   decay_steps=len(train_dataset) * 5, 
                                                   decay_rate=0.5, 
                                                   staircase=True)
    
    model.compile(optimizer=optimizers.Adam(learning_rate=lr_sch), # learning rate 스케쥴러를 사용해 상수 대신 설정.
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    
    return model
```

```python
model_lr = get_model_lr(LEARNING_RATE)
hist = model_lr.fit(train_dataset, epochs=N_EPOCH, validation_data=test_dataset)
```

```python
plot_result(hist)
```

## callback 을 이용해 학습률 변경
- ReduceLROnPlateau callback을 이용
- 로그를 남길 수 있어 변경되는 것을 확인할 수 있다.

```python
def get_model_lr2(initial_lr=0.01):
    model = keras.Sequential()
    model.add(layers.Flatten(input_shape=(28,28)))
    model.add(layers.Dense(256, activation='relu'))
    model.add(layers.Dense(128, activation='relu'))
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(10, activation='softmax', name='output_layer'))
    
    model.compile(optimizer=optimizers.Adam(learning_rate=initial_lr),
                  loss='categorical_crossentropy', metrics=['accuracy'])
    
    return model
```

```python
model_lr2 = get_model_lr2(LEARNING_RATE)
```

```python
import tensorflow as tf
from tensorflow import keras

# callback -> tensorflow.keras.callbacks 모듈의 클래스로 구현되어있음.
## 필요한 callback 클래스의 객체를 생성하여 fit() 호출 시 전달. (학습할 때 전달.)

### ReduceLROnPlateau : 학습 시 더 이상 성능 개선이 안되면, learning rate를 변경 (only 줄인다.)
lr_callback = keras.callbacks.ReduceLROnPlateau(monitor='val_loss', # 성능개선과 관련해 볼 평가 지표. 
                                                                    # default : val_loss 
                                                
                                                factor=0.5, # LR 조정 비율 (현재 LR * factor)
                                                
                                                patience=3, # 성능 개선 안될 때, 지정한 patience 수 만큼의
                                                            # epoch 만큼 기다린다. 
                                                # 보통 5~10 정도 줌, 여기선 30 epoch 돌릴거라 3으로 지정.
                                                
                                                verbose=2  # LR 변경될 때 로그를 남긴다.
                                               )

### ==> val_los 가 3에폭 동안 성능 개선이 안되면, 현재 LR에 0.5를 곱한 값을 새로운 LR로 변경하여 진행





#### 성능이 개선될 때 마다 모델을 저장하는 callback 생성.
save_dir = r"C:\workspace\Bootcamp_DeepLearning\saved_model\mnist_best_model"

mc_callback = keras.callbacks.ModelCheckpoint(filepath = save_dir, # 저장할 디렉토리
                                          save_best_only = True, # 성능이 개선될 때마다 저장한다.
                                          monitor = "val_loss", # default 값이므로 생략 가능, 성능개선평가지표
                                          verbose = 2, # 저장할 때마다 로그(기록) 출력
)
### => val_loss 성능이 개선될 때마다 모델을 save_dir 에 저장한다.

```

#### ctrl + shift + - => 셀 나눔

```python
hist = model_lr2.fit(train_dataset,
                     epochs=N_EPOCH+10, # 에폭 30으로 늘림.
                     validation_data=test_dataset, 
                     callbacks=[lr_callback, mc_callback] 
                                         # callback 을 여러 개 등록할 수 있으므로, 객체를 리스트로 전달함.
)
```

```python
saved_model = keras.models.load_model(save_dir) # 가장 성능 좋았을 때 저장된 모델
```
